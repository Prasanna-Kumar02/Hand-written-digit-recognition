### Image Conversion into Pixel values [Dataset - 40x40 pixels & 0-White 255-Black]

#### Train Set

```python
#Conversion of images into pixelvalues and export as csv file for digits

import os #To change directory and retrieve file names
import pandas as pd #to store values in data frame and export to csv
from PIL import Image #to convert the image into grayscale and generate pixel values

#Changing the root directory
os.chdir("C:/Users/Sanjeev Varma/Desktop/Capstone/Signature Recognition/Data") 

#Retrieving all the filenames in root directory and storing them in a list
files = os.listdir("C:/Users/Sanjeev Varma/Desktop/Capstone/Signature Recognition/Numbers")

#Creating empty list for storing pixel values
digits = list()
lst = list()

#Opening images in each file in the directory
for names in files:
    #getting the file path for each image
    path = "C:/Users/Sanjeev Varma/Desktop/Capstone/Signature Recognition/Numbers/"+names
    digit_name = os.listdir(path)
    for number in digit_name:
        digpath = "C:/Users/Sanjeev Varma/Desktop/Capstone/Signature Recognition/Numbers/"+names+"/"+number
        
        #Converting each image into 8-bit grayscale
        img = Image.open(digpath).convert('L')
        #Resizing image from 100x100 pixel to 40x40 pixel
        imgresize = img.resize((40,40),Image.ANTIALIAS)
        #Converting image data to a list of pixel values
        data = list(imgresize.getdata()) 
        #General representation is 255-white and 0-black in grayscale.
        #We are inversing this to 0 as white and 255 as black
        image = [(255 - x)   for x in data]
        
        #Storing the values and labels into list
        digits.append(names)
        for x in image:
            digits.append(x)
        lst.append(digits)
        digits = []
#Storing the lists as a dataframe and saving as csv file       
df = pd.DataFrame(lst)
print(df)
df.to_csv('train_numbers.csv',index=False,header=False)
```

### Image Conversion into Pixel values [Dataset - 40x40 pixels & 255-White 0-Black]

#### Train Set

```python
#Conversion of images into pixelvalues and export as csv file for digits
import os #To change directory and retrieve file names
import pandas as pd #to store values in data frame and export to csv
from PIL import Image #to convert the image into grayscale and generate pixel values

#Changing the root directory
os.chdir("C:/Users/Sanjeev Varma/Desktop/Capstone/Signature Recognition/Data")

#Retrieving all the filenames in root directory and storing them in a list
files = os.listdir("C:/Users/Sanjeev Varma/Desktop/Capstone/Signature Recognition/Numbers")

#Creating empty list for storing pixel values
digits = list()
lst = list()

#Opening images in each file in the directory
for names in files:
    #getting the file path for each image
    path = "C:/Users/Sanjeev Varma/Desktop/Capstone/Signature Recognition/Numbers/"+names
    digit_name = os.listdir(path)
    for number in digit_name:
        digpath = "C:/Users/Sanjeev Varma/Desktop/Capstone/Signature Recognition/Numbers/"+names+"/"+number
        #Converting each image into 8-bit grayscale
        img = Image.open(digpath).convert('L')  
        #Resizing image from 100x100 pixel to 40x40 pixel
        imgresize = img.resize((40,40),Image.ANTIALIAS)
        #Converting image data to a list of pixel values
        data = list(imgresize.getdata()) 
        #General representation is 255-white and 0-black in grayscale.
        #We are not inversing this to 0 as white and 255 as black  
        image = [x   for x in data]
        #Storing the values and labels into list
        digits.append(names)
        for x in image:
            digits.append(x)
        lst.append(digits)
        digits = []
#Storing the lists as a dataframe and saving as csv file         
df = pd.DataFrame(lst)
print(df)
df.to_csv('trainzero.csv',index=False,header=False)
```
